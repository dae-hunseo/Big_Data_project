```python
# API 키 불러오기
import json

with open("./secret.json", "r") as f:
    KEY = json.load(f)
```

```python
from urllib import request
from bs4 import BeautifulSoup

import pandas as pd

API = "https://api.odcloud.kr/api/15007122/v1/uddi:95d6cbf2-f800-4ce3-a4f7-f57823274732?"
```

```python
pageNum = 1
df2020 = pd.DataFrame()
print(f"page ", end=" ")

while True:
    print(pageNum, end=" ")
    url = API + f"serviceKey={KEY['ENCODING']}&perPage={1000}&page={pageNum}"
    
    result = request.urlopen(url)
    soup = BeautifulSoup(result, "html.parser")
    json_data = json.loads(str(soup))
    
    if json_data["currentCount"] == 0:
        break
    
    df2020 = pd.concat([df2020, pd.DataFrame(json_data["data"])], axis=0, ignore_index=True)
    
    pageNum = pageNum + 1
```

## 열 정리

```python
import pandas as pd

data = pd.read_csv("./건강검진_beforeclassify.csv", encoding="cp949")
```

```python
data[["기준년도", "시도코드"]].value_counts().sort_index()
```

```python
# 그대로 사용
data["SEX"] = data["성별코드"]
data["HEIGHT"] = data["신장(5Cm단위)"]
data["WEIGHT"] = data["체중(5Kg 단위)"]
data["ALCOHOL"] = data["음주여부"]
```

```python
# replace
region = {11: "서울", 26: "부산", 27: "대구", 28: "인천", 29: "광주", 
          30: "대전", 31: "울산", 36: "세종", 41: "경기", 42: "강원", 
          43: "충북", 44: "충남", 45: "전북", 46: "전남", 47: "경북", 
          48: "경남", 49: "제주", 50: "제주"}

data["지역"] = data["시도코드"].replace(region)
```

```python
# 1 condition
data["DIABETE"] = data["식전혈당(공복혈당)"].apply(lambda x: 1 if x >= 126 else 0)
data["CHOLESTEROL"] = data["총 콜레스테롤"].apply(lambda x: 1 if x > 250 else 0 if pd.notna(x) else x)
data["SMOKE"] = data["흡연상태"].apply(lambda x: 1 if x == 2 or x == 3 else 0)
```

```python
data["KIDNEY"] = data.apply(lambda x: 1 if x["요단백"]>=3 or x["혈청크레아티닌"]>1.7 
                            or x["혈청크레아티닌"]<0.8 else 0, axis=1)
data["KIDNEY"].value_counts()
```

```python
data["KIDNEY"] = data.apply(lambda x: 1 if x["요단백"]>=3 and (x["혈청크레아티닌"]>1.7 
                            or x["혈청크레아티닌"]<0.8) else 0, axis=1)
data["KIDNEY"].value_counts()
```

```python
# more than 2 conditions
data["BLIND"] = data.apply(lambda x: 1 if x["시력(우)"]==9.9 or x["시력(좌)"]==9.9 else 0, axis=1)
data["DEAF"] = data.apply(lambda x: 1 if x["청력(우)"]==2 or x["청력(좌)"]==2 else 0, axis=1)
data["HYPERTENSION"] = data.apply(lambda x: 1 if x["이완기 혈압"]>=90 or x["수축기 혈압"]>=140 else 0, axis=1)
data["LIVER_1"] = data.apply(lambda x: 1 if x["(혈청지오티)ALT"]>40 or x["(혈청지오티)AST"]>40 else 0, axis=1)
data["LIVER_2"] = data.apply(lambda x: 1 if (x["SEX"]=="2" and x["감마 지티피"]>35) 
                            or (x["SEX"]=="1" and x["감마 지티피"]>63) else 0, axis=1)
data["LIVER"] = data.apply(lambda x: 1 if x["LIVER_1"]==1 or x["LIVER_2"]==1 else 0, axis=1)
```
